# Using K-Means to Interpret the Data

```python
from sklearn.cluster import KMeans
import numpy as np

income_happy = np.column_stack((income, happy))
# print(income_happy)

km_res = KMeans(n_clusters = 10).fit(income_happy)

clusters = km_res.cluster_centers_
plt.scatter(income, happy)
plt.scatter(clusters[:, 0], clusters[:, 1], s = 2000, alpha = 0.3)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from sklearn.cluster import KMeans
```

```python
#loading a CSV file
data_frame = pd.read_csv('happyscore_income.csv')  
data_frame
#I will sort/filter data using by where avg_income > 15000
richest = data[ data['avg_income'] > 15000]
#insertinh happy score and avg_income, I want show indication of happiniest
# filtering happy is the column from data richest
happy = richest['happyScore']
income = richest['avg_income']
# I will insert income_inequality for show information about income_inequality
ineq = richest['income_inequality']
#creating stack for KMeans algorithm
income_happy = np.column_stack((income, happy))
# KMeans algorithm for define centroid, and I chose 3 center, k = 3
# first group is pooriest, 2 - normal, 3 - richest 
km_res = KMeans(n_clusters = 3).fit(income_happy)
cluster = km_res.cluster_centers_
# using matplotlib I'll show graph happiness, average income
fig, graph = plt.subplots()
# xlabel I will call like "income", ylabel will be "happy score"
plt.xlabel('income')
plt.ylabel('happy score')
#the balls size will be ineq*6, and alpha  = 0.42 
graph.scatter(income, happy, s = ineq * 6, alpha = 0.42)
#I took s = 6000 and alpha = 0.32 for show centroids
graph.scatter(cluster[:, 0], cluster[:, 1], s = 6000, alpha = 0.32)
# using this algorithm we can show country nome
for k, row in richest.iterrows():
    graph.text(row['avg_income'], 
             row['happyScore'], 
             row['country'])
print(graph)
```

# Project task

## About data : 
Dataset about distinguishing genuine and forged banknotes. Data were extracted from images that were taken from genuine and forged banknote-like specimens. For digitization, an industrial camera usually used for print inspection was used. The final images have 400x 400 pixels. Due to the object lens and distance to the investigated object gray-scale pictures with a resolution of about 660 dpi were gained. A Wavelet Transform tool was used to extract features from these images.

Attribute Information
V1. variance of Wavelet Transformed image 
V2. skewness of Wavelet Transformed image 

```python
# liberaries
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import pandas as pd
from sklearn.cluster import KMeans
```

```python
# exploration Loading the given Banknote authentication dataset.
data_banknote = pd.read_csv('Banknote authentication dataset.csv')
# geting right data columns
v1 = data_banknote['V1']
v2 = data_banknote['V2']
# normalisation of data
min_v1 = np.min(v1)
max_v1 = np.max(v1)
min_v2 = np.min(v2)
max_v2 = np.max(v2)

# print(min_v1, min_v2)
# print(max_v1, max_v2)

normalized_v1 = (v1 - min_v1)/(max_v1 - min_v1)
normalized_v2 = (v2 - min_v2)/(max_v2 - min_v2)
# Calculating statistical measures, e.g. mean and standard deviation.
mean = np.mean(data_banknote)
# print(mean)
std = np.std(data_banknote)
# print(std)
data = np.array(data_banknote)
# Visualising my data as my consider fit. 
fig, graph = plt.subplots()
graph.scatter(normalized_v1, normalized_v2)
plt.title('Banknote authentication')
plt.xlabel('V1')
plt.ylabel('V2')

#Modelling

v1_v2 = np.column_stack((normalized_v1, normalized_v2))
# using KMeans algorithm
km_res = KMeans(n_clusters = 2).fit(v1_v2)
cluster = km_res.cluster_centers_
graph.scatter(cluster[:, 0], cluster[:, 1], s = 1000, alpha = 0.5)
print('1 time:', cluster[:, 0], cluster[:, 1])
km_res = KMeans(n_clusters = 2).fit(v1_v2)
cluster = km_res.cluster_centers_
graph.scatter(cluster[:, 0], cluster[:, 1], s = 500, alpha = 0.5)
print('2 time:', cluster[:, 0], cluster[:, 1])
km_res = KMeans(n_clusters = 2).fit(v1_v2)
cluster = km_res.cluster_centers_
graph.scatter(cluster[:, 0], cluster[:, 1], s = 250, alpha = 0.5)
print('3 time:', cluster[:, 0], cluster[:, 1])
km_res = KMeans(n_clusters = 2).fit(v1_v2)
cluster = km_res.cluster_centers_
graph.scatter(cluster[:, 0], cluster[:, 1], s = 50, alpha = 0.5)
print('4 time:', cluster[:, 0], cluster[:, 1])
```
