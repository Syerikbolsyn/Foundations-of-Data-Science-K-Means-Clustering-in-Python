```python
# liberaries
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import pandas as pd
from sklearn.cluster import KMeans
```

```python
# exploration Loading the given Banknote authentication dataset.
data_banknote = pd.read_csv('Banknote authentication dataset.csv')
# geting right data columns
v1 = data_banknote['V1']
v2 = data_banknote['V2']
# normalisation of data
min_v1 = np.min(v1)
max_v1 = np.max(v1)
min_v2 = np.min(v2)
max_v2 = np.max(v2)

# print(min_v1, min_v2)
# print(max_v1, max_v2)

normalized_v1 = (v1 - min_v1)/(max_v1 - min_v1)
normalized_v2 = (v2 - min_v2)/(max_v2 - min_v2)
# Calculating statistical measures, e.g. mean and standard deviation.
mean = np.mean(data_banknote)
# print(mean)
std = np.std(data_banknote)
# print(std)
data = np.array(data_banknote)
# Visualising my data as my consider fit. 
fig, graph = plt.subplots()
graph.scatter(normalized_v1, normalized_v2)
plt.title('Banknote authentication')
plt.xlabel('V1')
plt.ylabel('V2')

# using KMeans algorithm
v1_v2 = np.column_stack((normalized_v1, normalized_v2))

km_res = KMeans(n_clusters = 2).fit(v1_v2)
cluster = km_res.cluster_centers_
graph.scatter(cluster[:, 0], cluster[:, 1], s = 1000, alpha = 0.5)
print('1 time:', cluster[:, 0], cluster[:, 1])
km_res = KMeans(n_clusters = 2).fit(v1_v2)
cluster = km_res.cluster_centers_
graph.scatter(cluster[:, 0], cluster[:, 1], s = 500, alpha = 0.5)
print('2 time:', cluster[:, 0], cluster[:, 1])
km_res = KMeans(n_clusters = 2).fit(v1_v2)
cluster = km_res.cluster_centers_
graph.scatter(cluster[:, 0], cluster[:, 1], s = 250, alpha = 0.5)
print('3 time:', cluster[:, 0], cluster[:, 1])
km_res = KMeans(n_clusters = 2).fit(v1_v2)
cluster = km_res.cluster_centers_
graph.scatter(cluster[:, 0], cluster[:, 1], s = 50, alpha = 0.5)
print('4 time:', cluster[:, 0], cluster[:, 1])
```
